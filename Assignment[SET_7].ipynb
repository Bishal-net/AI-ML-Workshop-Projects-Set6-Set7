```python
import pandas as pd

# Load the dataset
file_path = "/content/SET-7 xl sheet.xlsx"
df = pd.read_excel(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display missing values per column
missing_values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Step 2: Handle missing values
df_cleaned = df.dropna()  # Drop rows with missing values if any

# Prepare features and labels
X = df_cleaned[['ground_clearance', 'height']]  # Features
y = df_cleaned['category']  # Labels

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Step 3: 70%-30% train-test split
X_train_70, X_test_30, y_train_70, y_test_30 = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

# Train classifier
model_70 = DecisionTreeClassifier()
model_70.fit(X_train_70, y_train_70)
y_pred_30 = model_70.predict(X_test_30)

# Step 4: 80%-20% train-test split
X_train_80, X_test_20, y_train_80, y_test_20 = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train classifier
model_80 = DecisionTreeClassifier()
model_80.fit(X_train_80, y_train_80)
y_pred_20 = model_80.predict(X_test_20)

# Step 5: Display accuracy
accuracy_70_30 = accuracy_score(y_test_30, y_pred_30)
accuracy_80_20 = accuracy_score(y_test_20, y_pred_20)

accuracy_70_30, accuracy_80_20
```
